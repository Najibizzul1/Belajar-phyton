```python
#inisiasi awal.
#import pandas dan mengaliaskan menjadi pd.
#menginisiasi df untuk membaca file csv dan mengubah formatnya menjadi Datetime
import pandas as pd
df = pd.read_csv('Sample - Superstore.csv', parse_dates=['Order Date', 'Ship Date'])
df.head()
```

```python
#melihat tipe data
df.dtypes
```

```python
#melihat data unik di kolom postal code
df['Postal Code'].unique()
```

```python
#menghitung total data unik pada kolom postal code
df['Postal Code'].nunique()
```

```python
#menghitung total data unik pada kolom postal code, region, dan category
df[['Postal Code', 'Region', 'Category']].nunique()
```

```python
#melihat data unik pada kolom postal code, region, dan category
for col in ['Postal Code', 'Region', 'Category']:
  print(f"{col}:")
  print(df[col].unique())
  print()
```

```python
#melihat data null setiap kolom ada berapa
(df.isnull().sum())
```

```python
#menghilangkan data duplikat
(df.duplicated().sum())
df.drop_duplicates(inplace=True)
```

```python
#melihat data penjualan per kategori dan membuat fungsi penjualan_perkategori
penjualan_perkategori = df.groupby('Category')['Sales'].sum()
penjualan_perkategori.sort_values(ascending=False)
```

```python
#melihat data barang per kategori
for category, data in df.groupby('Category'):
    print(f"Kategori: {category}")
    top_produk = (
        data.groupby('Product Name')['Sales']
        .sum()
        .sort_values(ascending=False)
    )
    print(top_produk)
    print()
```

```python
#melihat data penjualan per bulan dan membuat fungsi penjualan_perbulan
penjualan_perbulan = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()
top_5bulan = penjualan_perbulan.nlargest(1).index
print(top_5bulan)
print(penjualan_perbulan.nsmallest(5))
```

```python
#melihat data penjualan per barang dan membuat fungsi penjualan_perbarang
penjualan_perbarang = df.groupby('Product Name')['Sales'].sum()
top_produk = penjualan_perbarang.nlargest(5)
print(top_barang)
print(penjualan_perbarang.nsmallest(5))
```

```python
#melihat data penjualan barang pada bulan terbaik
df_tb = df[df['Order Date'].dt.to_period('M').isin(top_5bulan)]
df_tb.groupby([df_tb['Order Date'].dt.to_period('M'), 'Product Name'])['Sales'].sum().sort_values(ascending=False)
```

```python
#import matplotlib.pyplot dan mengaliaskan menjadi plt.
#melihat chart penjualan per kategori, bulan, dan juga produk.
import matplotlib.pyplot as plt
penjualan_perkategori.plot(kind='bar')
plt.title("jumlah penjualan per kategori")
plt.show()
print()
plt.figure(figsize=(20,6))
penjualan_perbulan.plot(kind='bar')
plt.title("jumlah penjualan per Bulan")
plt.show()
print()
plt.figure(figsize=(20,6))
top_produk.plot(kind='bar')
plt.title("jumlah penjualan per produk")
plt.show()
```
